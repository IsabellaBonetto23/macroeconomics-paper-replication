# Abstract

High-frequency changes in interest rates around FOMC announcements are a key way to study how monetary policy affects asset prices and the overall economy. However, recent research has questioned whether these monetary policy surprises are truly random and valid for measuring their impact on the economy. One concern is that these surprises might be related to macroeconomic and financial information already available before the FOMC meetings.

This paper tackles these issues in two ways. First, it broadens the set of monetary policy events to include speeches by the Federal Reserve Chair, which increases both the number and importance of policy announcements. Second, it investigates how predictable these monetary policy surprises are through the “Fed response to news” channel and adjusts them to remove predictable components based on earlier data.

Reassessing the effects of monetary policy, the study finds two main results:

1) The high-frequency effects on asset prices are mostly unchanged.

2) The estimated effects on the macroeconomy are much larger and more significant than those found in previous studies
   using high-frequency data.

# 1 Introduction

Over the last twenty years, economists have widely used high-frequency changes in interest rates around the Federal Reserve’s Federal Open Market Committee (FOMC) announcements—commonly called *monetary policy surprises*—to study how monetary policy affects asset prices and the broader economy. 

These short-term interest rate movements capture how financial markets react immediately to new policy information revealed during FOMC meetings.

The appeal of using monetary policy surprises lies in their ability to isolate *causal* effects. Because they measure financial market reactions in a very short window of time—often just a few minutes or hours—around the official announcement, they avoid many endogeneity problems that affect lower-frequency data. For example, FOMC decisions are typically finalized well before they are made public. Therefore, the central bank cannot have been reacting to real-time movements in financial markets during that short interval. Any change in asset prices that occurs immediately after the announcement can be confidently interpreted as a reaction *to* the monetary policy news, rather than a cause *of* it.

In contrast, at lower frequencies (for instance, monthly or quarterly data), it becomes much harder to separate cause and effect. Movements in interest rates and asset prices may both respond to underlying economic conditions, blurring the direction of causality. High-frequency identification thus provides a cleaner, more credible way to measure the true impact of monetary policy decisions on financial markets and the macroeconomy.


# Understanding High-Frequency Interest Rate Changes

When researchers talk about *high-frequency interest rate changes* around FOMC announcements, they are **not referring to the Federal Reserve’s official policy rate itself changing minute by minute**. The federal funds rate — the main policy rate — is indeed *set* by the central bank and changes only when the FOMC announces a new target.

Instead, high-frequency interest rate changes refer to **movements in financial market instruments that reflect expectations about future monetary policy**, measured over very short time intervals — typically within a window of **30 minutes before and after an FOMC announcement**.  

These instruments include:
- **Federal funds futures**
- **Eurodollar futures**
- **Treasury yields**, especially short-term maturities (like 3-month or 2-year yields)

Before an FOMC announcement, financial markets already have expectations about what the central bank will do. When the announcement is released, traders quickly update those expectations, and this adjustment causes interest rates in these markets to move — sometimes within seconds.

The *change* in these market rates immediately after the announcement captures how much the news surprised investors.  
- If the Fed raises rates more than expected, short-term yields jump.  
- If the Fed is more dovish (less aggressive) than expected, yields fall.  

Therefore, these **high-frequency rate movements are a measure of the “monetary policy surprise”** — the new information about the stance of policy that markets learn at the announcement. Researchers use these surprises as *exogenous shocks* to estimate the causal effects of monetary policy on asset prices and macroeconomic variables. 

Technically speaking is like if we are isolating the structural shock of monetary policy that is unexpected and unpredictable by markets. 



# Exogeneity and Recent Challenges to the Use of Monetary Policy Surprises

Monetary policy surprises are traditionally viewed as **unpredictable** using any publicly available information prior to the FOMC announcement. This assumption is essential because, if markets could anticipate these surprises, traders would exploit this predictability to earn profits, and prices would adjust in advance — eliminating the element of surprise altogether.  

Under this reasoning, monetary policy surprises are considered **exogenous** with respect to all macroeconomic variables known before the announcement. In other words, they are assumed to be independent of the current economic environment and can therefore serve as a valid *instrument* for identifying the causal effects of monetary policy in empirical models such as structural VARs (SVARs) or local projections (LPs), as discussed in Stock and Watson (2018).

However, several recent studies have challenged this assumption of exogeneity.  
For instance, **Cieslak (2018)**, **Miranda-Agrippino and Ricco (2021)**, and **Bauer and Swanson (2021)** have found that monetary policy surprises are often **correlated with macroeconomic or financial data** that was already available before the FOMC announcement.  

Specifically, Bauer and Swanson (2021) report that such correlations can explain between **10% and 40%** of the variation in monetary policy surprises (R² = 0.10–0.40). This finding suggests that these surprises may not be entirely independent from prior economic conditions.  

If this is the case, the standard assumption that monetary policy surprises represent purely exogenous changes becomes questionable. As a result, the validity of many empirical studies that rely on these high-frequency surprises as instruments for identifying monetary policy shocks may be partially undermined.



# Further Questions About Exogeneity and Relevance

Building on the previous discussion, some additional studies further question the reliability of using high-frequency monetary policy surprises as valid instruments.

**Ramey (2016)** argues that the estimated macroeconomic effects of monetary policy are often weak or inconsistent for samples starting after the mid-1980s. One reason is that monetary policy since around 1984 has been conducted more systematically and predictably, reducing the number of large, unexpected policy shocks. As a result, the variation in structural monetary policy shocks—whether measured using high-frequency surprises or other identification methods—is smaller and less informative for the post-1984 period compared with earlier decades.

In simpler terms:
- Before 1984, monetary policy was less systematic, and surprises were larger and more informative.
- After 1984, the Federal Reserve’s behavior became more predictable, and true “surprises” became rarer, making it harder to estimate the effects of policy shocks accurately.

Combining this with previous findings:
- **Cieslak (2018)**, **Miranda-Agrippino and Ricco (2021)**, and **Bauer and Swanson (2021)** question the **exogeneity** of high-frequency monetary policy surprises (whether they are truly independent of prior information).
- **Ramey (2016)** questions their **relevance** (whether they contain enough variation to explain macroeconomic outcomes).

As noted by **Stock and Watson (2018)**, both *exogeneity* and *relevance* are essential for a variable to serve as a strong and valid instrument.  
If monetary policy surprises fail to meet either of these conditions, their usefulness for identifying the causal effects of monetary policy becomes limited.


# Addressing the Challenges: Improving Relevance and Exogeneity

In this paper, the authors respond to the recent critiques by addressing both the **relevance** and the **exogeneity** of monetary policy surprises.

### Improving Relevance

The first improvement focuses on making monetary policy surprises more **relevant** — that is, more informative about actual changes in monetary policy.  
To do this, the authors **expand the set of monetary policy announcements** beyond the traditional FOMC decisions to include:
- **Press conferences**
- **Public speeches**
- **Congressional testimony** by the Federal Reserve Chair

These events, are referred by the authors as “speeches” for brevity.
These "speeches" are shown by **Swanson and Jayawickrema (2021)** to have strong effects on financial markets, sometimes even greater than official FOMC statements. Including them therefore more than doubles the number of meaningful monetary policy events and enhances the informational content of the data.

By expanding the sample of announcements, the authors also **extend the time period back to 1988**, covering years when monetary policy was more variable and less predictable than in the post-1990s era. This responds directly to **Ramey (2016)**’s critique by increasing both the number and diversity of monetary policy shocks, making the dataset richer and more representative.

In summary:
- Broader coverage of policy communication events → higher *relevance*  
- Longer sample period (starting in 1988) → more variation and stronger identification power  

### Improving Exogeneity

The second improvement targets the **exogeneity** issue — the concern that monetary policy surprises might be correlated with prior macroeconomic or financial information.

Following **Bauer and Swanson (2021)**, the authors remove this predictable component by:
1. Regressing the high-frequency surprises on variables that capture economic and financial conditions known before each FOMC event.  
2. Taking the **residuals** from these regressions, which represent the *unpredictable* portion of the policy surprise.

These residuals — often called **orthogonalized surprises** — are purged of any correlation with pre-announcement information. This adjustment helps ensure that the remaining variation in monetary policy surprises is truly exogenous and not driven by market expectations or prior data.

As a result, these orthogonalized surprises can be used as cleaner instruments in econometric models such as **SVARs** or **local projections (LPs)**. They help eliminate potential biases, such as the so-called “price puzzle,” and yield more accurate estimates of how monetary policy affects macroeconomic variables.



# Why an Unpurified High-Frequency Instrument May Not Be Relevant or Exogenous

High-frequency monetary policy surprises (changes in interest rates around FOMC announcements) are often used as an **instrument** to identify monetary policy shocks.  
However, if this instrument — usually denoted by **zₜ** — is **not purified** (or “orthogonalized”) from other information that was already available before the announcement, it may fail to satisfy the two key requirements for a valid instrument: **relevance** and **exogeneity**.

---

## Relevance

**Definition:**  
The instrument must be correlated with the *true* monetary policy shock **ε₁,ₜ**:

$$
E[\varepsilon_{1,t}(z_t - \bar{z})] = \phi \neq 0
$$

**Problem:**  
If the unpurified high-frequency surprise mixes different types of news — for example:
- true policy shocks (unexpected rate changes), **and**
- the Fed’s communication about the economic outlook (“information effect”),

then the instrument may not move one-to-one with the actual monetary policy shock.

In that case, part of the observed change in interest rates reflects **information about the economy**, not a **policy action**.  
As a result, the correlation between $z_t$ and the true policy shock $ \varepsilon_{1,t} $ becomes weak or noisy, reducing **relevance**.

**Intuition:**  
Markets react not only to what the Fed *does*, but also to what it *says* about future economic conditions.  
If we don’t remove that predictable, information-driven component, our instrument may no longer measure pure monetary shocks.

---

## Exogeneity

**Definition:**  
The instrument must be uncorrelated with *other* structural shocks **ε₂,ₜ**:

$$
E[\varepsilon_{2,t}(z_t - \bar{z})] = 0
$$

**Problem:**  
If the unpurified instrument contains elements related to macroeconomic or financial conditions known before the FOMC meeting, it will be **correlated** with other shocks — such as:
- demand shocks,  
- productivity news, or  
- financial uncertainty.

In that case, $z_t$ will move systematically with variables like inflation forecasts or output expectations.  
This means it is *not independent* from the rest of the economy — violating the exogeneity condition.

---

## Why “Purification” (Orthogonalization) Solves This

To fix these issues, researchers **purify** the instrument by removing its predictable component.  
They regress the raw surprise $z_t$ on macroeconomic and financial variables known before the announcement, and use the **residuals** as the new, orthogonalized instrument:

$$
z_t^{*} = z_t - E[z_t \,|\, X_{t-1}]
$$

where $X_{t-1}$ are pre-announcement variables.

- This step **restores exogeneity**, because $z_t^{*}$ is uncorrelated with prior information.  
- It also **improves relevance**, because it isolates the part of $z_t$ that truly captures the unexpected policy component.

---

## Summary Table

| Property | What It Requires | Why It Fails for Unpurified $z_t$ | Fix |
|-----------|------------------|--------------------------------------|------|
| **Relevance** | $E[\varepsilon_{1,t}(z_t - \bar{z})] \neq 0$ | Mixed with Fed information → weak link to true shock | Keep only unexpected (policy) part |
| **Exogeneity** | $E[\varepsilon_{2,t}(z_t - \bar{z})] = 0$ | Correlated with known macro/financial data | Remove predictable component using pre-announcement info |

# Section 3.1 – Monetary Policy Surprises around FOMC Announcements

This section explains **how the authors construct the high-frequency monetary policy surprises (MPS)** used in their analysis.  
The goal is to capture **unexpected changes in monetary policy** around Federal Open Market Committee (FOMC) announcements — i.e., how financial markets react to *new* information about the Fed’s policy decisions.

---

## Data Sources and Coverage

The dataset is based on **Swanson and Jayawickrema (2021)**, covering the period from **1988 to 2019**.  
It includes:
- **322 FOMC announcements**, and  
- **880 speeches** by the Federal Reserve Chair (press conferences, testimonies, public remarks).

By including both FOMC announcements and Chair communications, the authors capture a much larger set of events when markets receive new information about monetary policy.

### Why after 1994?
Before 1994, the Fed did **not officially announce** its target rate decisions immediately.  
Markets had to infer policy changes indirectly by observing **open market operations**.  
From **1994 onward**, FOMC decisions began to be explicitly communicated via official press releases, giving researchers a clear event timestamp to study market reactions.

---

## Measuring High-Frequency Monetary Policy Surprises

To measure how markets react, researchers look at **changes in short-term interest rate futures** within a *very narrow time window* around each FOMC announcement.

Specifically, the change is measured over a **30-minute window**:
- starting **10 minutes before** the announcement,  
- and ending **20 minutes after** it.  

This ensures that only the market’s *immediate response* to the policy news is captured, excluding unrelated developments.

The data come from **Tick Data**, which records minute-by-minute transactions.

---

## The Eurodollar Futures Market

### What is a Eurodollar future?

A **Eurodollar future** is a financial contract that reflects the **market’s expectation of the 3-month U.S. dollar LIBOR rate** (the interest rate on dollar deposits held outside the U.S.) for a given future date.

- Traded on the **Chicago Mercantile Exchange (CME)**.  
- Quoted as:
  $$
  \text{Price} = 100 - \text{Expected 3-month LIBOR rate}
  $$
  → If the price **falls**, markets expect **higher short-term interest rates** (tighter policy).

Because Eurodollar futures are extremely liquid and react within seconds to news,  
**changes in their prices around FOMC announcements** provide a clean measure of how expectations for future rates shift in response to new monetary policy information.

---

## Alternative High-Frequency Instruments Used in the Literature

Different studies have used **alternative financial instruments** to capture monetary policy surprises.  
The common goal across all approaches is to measure **unexpected changes in interest rate expectations** during small windows around FOMC events.

### Examples:

- **Federal Funds Futures (Kuttner, 2001):**  
  Measures changes in the expected *target* federal funds rate over the current month.

- **Eurodollar Futures (Gürkaynak, Sack, and Swanson, 2005):**  
  Capture changes in the *expected path* of future short-term rates, beyond just the current target.

- **Treasury Yields (Cook and Hahn, 1989):**  
  Use changes in yields on short-maturity Treasury securities around FOMC announcements.

- **Interest Rate Swaps (Gertler and Karadi, 2015):**  
  Focus on swap rates to identify shocks to longer-term expectations.

Each market instrument reflects slightly different aspects of policy expectations:
- **Federal funds futures:** immediate, short-horizon expectations.  
- **Eurodollar futures:** near-term (3–12 months ahead) expectations.  
- **Treasuries or swaps:** broader expectations about the entire yield curve.

The authors of this paper choose Eurodollar futures because:
1. They are highly liquid and respond quickly to monetary news.  
2. They capture *both* short- and medium-term expectations.  
3. They provide a consistent time series across several decades.

---

## How the Surprise is Constructed

The authors focus on the **first four Eurodollar futures contracts (ED1–ED4)**, corresponding to maturities up to about one year ahead.  
They compute the **first principal component (PC1)** of the changes in these contracts:

$$
MPS_t = PC_1(\Delta ED1_t, \Delta ED2_t, \Delta ED3_t, \Delta ED4_t)
$$

where $\Delta ED_i$ is the change in the Eurodollar futures rate for contract $i$ in the 30-minute window around the FOMC announcement.

### Why use the first principal component?
Because it summarizes the common variation across the contracts:
- **Target factor:** immediate change in the federal funds rate target.  
- **Path factor:** change in the expected trajectory of rates over upcoming months.

The first principal component is a **weighted average** of both, capturing the core of the market’s reaction.

- **MPS_t** is a syntetic variable that reasume the overall movement in EURODOLLAR futures and so it is a proxy for the monetary policy surprise (unexpected structural shock). We will see that actually even this instrument can be predicted an so it does not represent the real unexpected shock of monetary policy. 
---

## Economic Intuition

- When the Fed **tightens policy more than expected**, Eurodollar futures **fall** (prices down → rates up).  
- When the Fed **eases** or signals lower rates, futures **rise**.

Thus, $MPS_t$ captures the **unexpected component** of monetary policy decisions — the *surprise* in the Fed’s stance relative to what markets had anticipated.

# Section 3.2 – Predictability with Macroeconomic and Financial Data

In this section, the authors test whether **monetary policy surprises (MPS)** can be predicted using macroeconomic and financial information available **before** each FOMC announcement.  
If the surprises are partly predictable, they may not represent purely exogenous monetary policy shocks.

---

## Objective of the Analysis

The goal is to verify whether the high-frequency monetary policy surprises — measured from changes in Eurodollar futures around FOMC announcements — are **truly unpredictable ex ante**.

If markets could systematically forecast part of these surprises using public information, then the MPS would be correlated with variables known before the announcement, violating **exogeneity**.

---

## Predictive Regression Framework

The authors estimate the following regression:

$$
mps_t = \alpha + \beta X_{t-} + u_t
$$

where:
- $mps_t$ is the monetary policy surprise for FOMC announcement $t$;  
- $X_{t-}$ is a vector of macroeconomic and financial variables observed *before* the announcement;  
- $u_t$ is the residual (unpredictable part).

If the estimated coefficients $\beta$ are **statistically significant**, it means that the MPS can be partially predicted using these variables — i.e., they are **not fully exogenous**.

---

## Predictors Used

The authors use **six predictors**, all measured before each FOMC announcement, chosen because they have an intuitive economic link to the Fed’s policy decisions.

### 1️⃣ Nonfarm Payrolls Surprise  
- Measures the difference between the released payrolls data and the market’s expected value (from Money Market Services survey).  
- A positive surprise in payrolls (stronger labor market) usually leads to **tighter** monetary policy expectations.

### 2️⃣ Employment Growth (12 months)  
- Annual change in nonfarm employment before the announcement (Cieslak, 2018).  
- Reflects broader labor market conditions that may influence the Fed’s stance.

### 3️⃣ S&P 500 Change (3 months)  
- Log change in the S&P 500 index over the 3 months prior to the FOMC meeting.  
- Rising equity markets indicate optimism and stronger demand, possibly leading the Fed to tighten policy.

### 4️⃣ Yield Curve Slope (3 months)  
- Change in the slope of the yield curve (10-year minus 2-year zero-coupon Treasury yields).  
- A flatter or inverted curve suggests market expectations of future easing; a steeper curve often signals tightening.

### 5️⃣ Commodity Prices (3 months)  
- Log change in the Bloomberg Commodity Spot Price Index.  
- Higher commodity prices suggest inflationary pressures, leading to tighter monetary policy.

### 6️⃣ Treasury Skewness  
- Implied skewness of 10-year Treasury note futures (option-implied), averaged over the preceding month.  
- Captures the balance of risks perceived in bond markets — whether investors expect higher or lower rates.

---

## Empirical Results

The regression results (Table 1 in the paper) show the following pattern:

| Predictor | Sign of Coefficient | Interpretation |
|------------|--------------------|----------------|
| Nonfarm Payrolls | **Positive** | Strong labor data → tighter expected policy |
| Employment Growth | **Positive** | Robust job market → higher likelihood of tightening |
| S&P 500 | **Positive** | Rising equities → Fed more likely to raise rates |
| Yield Curve Slope | **Negative** | Steeper curve → easing cycle likely ending |
| Commodity Prices | **Positive** | Inflation pressure → Fed tightening |
| Treasury Skewness | **Positive** | Higher skewness → greater tightening expectations |

The $R^2$ across models ranges from **0.16 to 0.17**, meaning that about **16–17%** of the variation in MPS can be explained by these pre-announcement variables.

This confirms **a moderate but statistically significant degree of predictability** in monetary policy surprises.

---

## Robustness and Alternative Specifications

The authors test the regression across multiple subsamples:
- **1988–2019 (full sample)**  
- **1994–2019** (period of explicit FOMC communication)  
- **1988–2007** (pre-financial crisis)  
- **1990–2019** (extended sample including FF4 data from Bauer & Swanson, 2021)

Across all periods, results remain consistent — both in sign and significance — confirming the robustness of the findings.

---

## Interpretation and Economic Intuition

The key takeaway is that **MPS are not entirely random**.  
They appear to reflect **systematic reactions of the Fed to economic conditions** that are already observable before meetings.

For example:
- Strong employment or inflation data → markets expect higher rates → smaller “surprise” when the Fed tightens.  
- Weak growth or falling markets → markets anticipate easing → smaller or negative surprises.

This evidence aligns with the theoretical model from Section 2:  
predictability arises because the **private sector learns about the Fed’s policy rule** over time and uses public information to form expectations.

---

## Ex Ante vs. Ex Post Predictability

Finally, the authors emphasize a crucial distinction:
- **Ex ante predictability:** whether investors could have predicted MPS *in real time* before announcements.  
- **Ex post predictability:** correlation observed after the fact, when all data are known.

Their model implies that MPS are **not predictable ex ante**, because market participants have imperfect information about the Fed’s reaction function.

They test this by estimating **out-of-sample regressions** (expanding window forecasts) and find:
- Out-of-sample $R^2$ is **close to zero**.  
- Forecast gains from the six predictors are **not statistically significant**.

This confirms the absence of true **ex ante predictability** — even though MPS appear correlated with economic data *ex post*.

# Sections 3.3–3.4 – Predictability with Macroeconomic Forecast Data and Interpretation

After analyzing how high-frequency monetary policy surprises (MPS) can be predicted using macroeconomic and financial variables, the authors investigate whether such predictability is driven by **information contained in macroeconomic forecasts**.  
They then interpret the overall evidence on what this predictability means for the exogeneity of monetary policy shocks.

---

## Motivation: The Role of Macroeconomic Forecasts

In a recent and influential study, **Miranda-Agrippino and Ricco (2021)** (MAR) found that the Fed’s internal **Greenbook forecasts** are **strongly correlated** with high-frequency monetary policy surprises.  
This finding suggested that when the Fed releases new policy decisions, the surprises may partly reflect **private information** about the Fed’s internal economic outlook.

If true, this would mean that monetary policy surprises are **not purely exogenous**, but instead incorporate information about how the Fed interprets the economy.

The authors of this paper reexamine this claim using **publicly available forecasts**, such as the **Blue Chip survey**, to see whether similar correlations exist even without access to the Fed’s private information.

---

## Empirical Strategy

They replicate the MAR predictive regressions, following **Romer and Romer (2004)**, by regressing the high-frequency monetary policy surprise on **macroeconomic forecast variables** known before the announcement:

$$
mps_t = \alpha + \beta F_{t-} + u_t
$$

where:
- $mps_t$ is the high-frequency monetary policy surprise,
- $F_{t-}$ is a vector of forecast data (either Greenbook or Blue Chip forecasts),
- $u_t$ is the residual term.

The forecasts include:
- Real GDP growth (current and up to three quarters ahead),
- GDP deflator inflation,
- Unemployment rate forecasts (current and up to two quarters ahead).

These are the same forecast horizons as in **Miranda-Agrippino and Ricco (2021)**, allowing for a direct comparison.

---

## Greenbook vs. Blue Chip Forecasts

### Greenbook Forecasts:
- Internal forecasts prepared by the Federal Reserve staff before each FOMC meeting.
- Not publicly available at the time of the meeting (released with a five-year lag).
- Represent the Fed’s **private view** of the economy.

### Blue Chip Forecasts:
- Compiled from a panel of professional private-sector forecasters.
- Publicly available before each FOMC meeting.
- Reflect the **market’s consensus** expectations for key macroeconomic variables.

---

## Results

### Regression Results
The results, reported in Table 2 of the paper, show:

| Forecast Source | R² Range | Interpretation |
|------------------|----------|----------------|
| **Greenbook forecasts** | 0.16 – 0.23 | Moderate predictability |
| **Blue Chip forecasts** | 0.14 – 0.22 | Similar predictability |

The main finding is that **predictability is nearly identical** whether one uses the Fed’s Greenbook forecasts or the public Blue Chip forecasts.

### Implication:
- The predictive content of the Fed’s internal forecasts is **not unique**.  
- Public forecasts contain roughly the same information as the Fed’s own internal outlook.

Therefore, the correlation between forecasts and monetary policy surprises likely reflects **publicly available information**, not private Fed information.

---

## Interpretation of the Predictability Evidence

The authors then discuss how to interpret this broader evidence on the predictability of monetary policy surprises.

### (a) Ex Post Predictability Does Not Imply Ex Ante Forecastability
The high-frequency surprises are **unpredictable ex ante**, meaning market participants could not foresee them in real time.  
However, **ex post**, when all information is revealed, we observe correlations because the Fed reacts systematically to the same public data that markets also observe.

This is consistent with **learning models** (e.g., Timmermann, 1993) where agents form expectations based on imperfect and evolving information.

---

### (b) Rejecting the "Private Fed Information" Hypothesis
While **Miranda-Agrippino and Ricco (2021)** interpreted the correlation between MPS and Greenbook forecasts as evidence of **private Fed information**,  
the authors of this paper argue that the similarity between Greenbook and Blue Chip results suggests otherwise.

In other words:
> The predictability of MPS arises not from special Fed insights, but because both the Fed and the private sector observe and react to the same macroeconomic signals.

This interpretation aligns with **Bauer and Swanson (2021)**, who also found that public and private forecasts are statistically indistinguishable in terms of predictive power.

---

### (c) Connection to the Learning Model
The results are consistent with the **learning model** presented earlier (Section 2), in which:
- The Fed’s policy responses depend on macroeconomic conditions that are also visible to the private sector.
- Market participants gradually learn about the Fed’s policy rule over time.
- Hence, observed predictability does not mean the Fed has superior private information — it means **both sides share the same data**, but the Fed’s reaction strength is uncertain.

---

## Broader Interpretation

The authors caution that while MPS appear predictable using macroeconomic and forecast data, this **does not invalidate their use** as a measure of monetary policy shocks — provided the predictable component is properly removed.

Purifying (orthogonalizing) the high-frequency surprises ensures that:
- The residual component reflects **unexpected** policy changes,
- The instrument remains **relevant** and **exogenous** for structural analysis (as discussed earlier).

---

## Summary

| Topic | Finding | Implication |
|--------|----------|-------------|
| Predictability using Fed forecasts | Yes, but moderate | Driven by public info, not private insights |
| Predictability using Blue Chip forecasts | Nearly identical | Confirms results based on Greenbook |
| Ex post vs. ex ante | Ex post correlation, no ex ante forecastability | Consistent with imperfect information learning |
| Policy implication | MPS must be orthogonalized | Ensures exogeneity in empirical applications |

# 4.2 – Conventional and Orthogonalized Monetary Policy Surprises

## Background: Why the Standard MPS Is Not Truly Exogenous

The **Monetary Policy Surprise (MPS)** is meant to capture **unexpected monetary policy shocks**, i.e., changes in expectations about future short-term interest rates that occur in a very short window around FOMC announcements.  
It is constructed from **high-frequency changes in Eurodollar futures** — contracts that reflect market expectations of future short-term interest rates.

However, as shown in Section 3, these high-frequency surprises are **predictable** to some extent using macroeconomic and financial data available **before** the FOMC announcement.  
This predictability indicates that the raw MPS is **not purely exogenous**, because part of its variation reflects information that was already publicly known before the policy decision.

### Example of the problem

Suppose the MPS increases because the market expected stronger economic growth or higher inflation.  
In that case, MPS is correlated with **macroeconomic conditions**, not only with the **true policy shock**.  
Formally, if $X_{t-}$ are macro/financial variables known before the FOMC meeting, then:

$$
E[MPS_t \,|\, X_{t-}] \neq 0
$$

This correlation violates the **exogeneity condition** required for using MPS as an instrument in structural analyses (like SVARs or Local Projections).  
It means that part of what we interpret as a "policy surprise" is actually a **systematic reaction** to public economic information — not a truly unexpected policy shock.

---

## Goal: Purifying the Instrument

To recover a valid measure of exogenous monetary policy shocks, the authors **orthogonalize** the MPS — that is, they **remove the predictable component** related to macroeconomic and financial variables observed before the announcement.

The goal is to obtain a “clean” version of the instrument, denoted $MPS_t^{\perp}$, that satisfies:

- **Relevance:** correlated with the true monetary policy shock;  
- **Exogeneity:** uncorrelated with any pre-announcement macro or financial information.

This orthogonalization ensures that $MPS_t^{\perp}$ represents **only the unexpected part** of the policy change.

---

## The Orthogonalization Procedure

The authors proceed in two steps:

### Step 1 – Predict the MPS
They first estimate a predictive regression using the macroeconomic and financial variables $X_{t-}$ identified in Section 3 (e.g., nonfarm payrolls, employment growth, S&P 500 changes, yield curve slope, commodity prices, and Treasury skewness):

$$
MPS_t = \alpha + \beta' X_{t-} + u_t
$$

Here, $u_t$ represents the unpredictable part of the monetary policy surprise.

### Step 2 – Compute the Residuals
Then they define the **orthogonalized MPS** as the residual from this regression:

$$
MPS_t^{\perp} = MPS_t - \hat{\alpha} - \hat{\beta}' X_{t-}
$$

This residual captures the portion of the MPS that is **uncorrelated** with the predictors known before the announcement.

Formally:

$$
E[MPS_t^{\perp} X_{t-}] = 0
$$

and

$$
E[MPS_t^{\perp}] = 0
$$

---

## Intuition and Interpretation

Economically, this procedure **“purifies”** the instrument by stripping out all the variation in the policy surprise that can be statistically explained by public information.  
What remains is a measure of the **true, unexpected monetary policy shock** — the part that financial markets and the private sector could not anticipate.

Thus:
- The raw $MPS_t$ includes both **expected** and **unexpected** components.  
- The orthogonalized $MPS_t^{\perp}$ isolates the **unexpected** component, making it a valid instrument for structural identification.

This directly addresses the concern that MPS might be endogenous due to correlation with economic conditions or expectations.

---

## Connection to the Instrumental Variable Conditions

For $MPS_t^{\perp}$ to be a valid instrument (say, $Z_t$), it must satisfy:

- **Relevance:**  
  $$
  E[\varepsilon_{1,t}(Z_t - \bar{Z})] = \phi \neq 0
  $$
  meaning that the purified instrument remains correlated with the true monetary policy shock.

- **Exogeneity:**  
  $$
  E[\varepsilon_{2,t}(Z_t - \bar{Z})] = 0
  $$
  meaning that it is uncorrelated with other structural shocks or pre-announcement information.

By orthogonalizing, we ensure that exogeneity is restored, while relevance is preserved because the residual still reflects true policy surprises.

---

## Final Outcome

After orthogonalization, the authors compare the effects of $MPS_t$ and $MPS_t^{\perp}$:
- On **asset prices** (within the event-study setup),
- And later (in Section 5) on **macroeconomic variables** using SVARs and Local Projections.

They find that:
- Asset price responses are very similar using either measure (suggesting financial markets react mainly to the truly unexpected part of policy changes).  
- But the macroeconomic effects are **much stronger and more significant** when using the **orthogonalized** version, confirming that purifying the instrument removes attenuation bias due to endogeneity.

---

## Key Takeaway

> The **orthogonalization of the Monetary Policy Surprise** transforms a noisy, partly predictable measure into a truly exogenous proxy for monetary policy shocks.  
> It ensures that the estimated effects of monetary policy reflect genuine causal responses, not mechanical correlations with pre-announcement macroeconomic conditions.

---

### Reduced-Form VAR Equation

The reduced-form VAR estimated by the authors is:

$$
Y_t = \alpha + A(L)Y_{t-1} + u_t
$$

where:
- $Y_t$ is a $(4 \times 1)$ vector containing the four endogenous variables  
  $\{ 2Y_t, \log(INDPRO_t), \log(CPI_t), EBP_t\}$,  
- $\alpha$ is a $(4 \times 1)$ vector of intercepts (constant terms),  
- $A(L)$ is a $(4 \times 4)$ matrix polynomial in the lag operator $L$,  
- $u_t$ is the $(4 \times 1)$ vector of reduced-form residuals, with $Var(u_t) = \Sigma_u$.

The polynomial $A(L)$ captures the dynamic relationships among the variables through **12 monthly lags**:

$$
A(L) = A_1 L + A_2 L^2 + \dots + A_{12} L^{12}
$$

Hence, the estimated equation can be written more explicitly as:

$$
Y_t = \alpha + A_1 Y_{t-1} + A_2 Y_{t-2} + \dots + A_{12} Y_{t-12} + u_t
$$

The reduced-form VAR is estimated using **monthly data from January 1976 to February 2020** via **ordinary least squares (OLS)**.  
The starting date is set to 1976 because the **2-year Treasury yield** series — one of the four variables in the VAR — is only available from that year onward.  
Although other series (such as industrial production or CPI) are available earlier, the consistent availability of all variables determines the beginning of the sample.

The sample ends in February 2020 to **avoid the extreme volatility** in industrial production and financial variables associated with the onset of the COVID-19 pandemic in the United States.

The model includes:
- A constant term ($\alpha$) to capture deterministic trends.  
- **12 monthly lags** ($A_1, A_2, \dots, A_{12}$), corresponding to one full year of macroeconomic dynamics.  
- Four endogenous variables in the vector $Y_t = \{ 2Y_t, \log(INDPRO_t), \log(CPI_t), EBP_t\}$.

All coefficients are estimated jointly by OLS for each equation in the VAR system.  
This setup matches the standard monthly VAR specifications used in monetary policy studies, while ensuring a balanced dataset over the full sample period (1976–2020).

## Estimating the Reduced-Form VAR(12)

We begin by estimating the **reduced-form Vector Autoregression (VAR)** with 12 lags, as in Bauer and Swanson.

Let:
- $m$ = number of variables in the VAR (here, $m = 4$),
- $p$ = number of lags (here, $p = 12$),
- $Y_t$ = the $m \times 1$ vector of macroeconomic variables at time $t$.

The reduced-form VAR(4) can be written as:

$$
Y_t = \alpha + A_1 Y_{t-1} + A_2 Y_{t-2} + A_3 Y_{t-3} +...+ A_{12} Y_{t-12} + u_t,
$$

where:
- $A_1, A_2, A_3,..., A_{12}$ are $m \times m$ coefficient matrices,
- $u_t$ is the $m \times 1$ vector of **reduced-form shocks**, which are in general **correlated contemporaneously** across variables.

---
## Reduced-Form VAR(12) Estimation (Manual, Matrix-Based)

Let the data matrix be:

$$
Y_t =
\begin{bmatrix}
\text{2Y\_TREASURY}_t \\
\text{INDPRO}_t \\
\text{CPI}_t \\
\text{2Y\_TREASURY}_t \\
\text{EXCESS BOND PREMIUM}_t
\end{bmatrix}
\qquad (4 \times 1)
$$

We lose the first 12 observations due to lagging.  
We stack all lagged variables to form the **regressor matrix**:

$$
X_t = \begin{bmatrix}
 1 & Y_{t-1}' & Y_{t-2}' & Y_{t-3}' & ... & Y_{t-12}'
\end{bmatrix}
\quad \text{(1 × 1+48)}.
$$

Stacking over time:

- $Y$ becomes a $(T - p) \times m$ matrix,
- $X$ becomes a $(T - p) \times (1+mp)$ matrix.

---

### Matrix Form Representation

To estimate the model, it is convenient to rewrite the VAR in **stacked matrix form**:

$$
Y = X \Phi + U,
$$

where:

| Symbol | Dimension | Meaning |
|--------|-----------|---------|
| $Y$ | $(T-p) \times m$ | Current observations of $Y_t$ (after losing the first $p$ rows due to lags) |
| $X$ | $(T-p) \times (mp)$ | Matrix of lagged variables |
| $\Phi$ | $(mp) \times m$ | Matrix collecting all VAR coefficients |
| $U$ | $(T-p) \times m$ | Matrix of reduced-form shocks |

The matrix $\Phi$ collects all lag coefficient matrices in one object:

$$
\Phi \text{ has dimension } (mp) \times m = (4 \cdot 12) \times 4 = 48 \times 9.
$$

Thus, **$\Phi$ contains all the information needed to describe the dynamics of the VAR.**  
This is why in code, the estimated coefficient matrix is denoted as:


---

### OLS Estimation

The reduced-form VAR is estimated equation-by-equation using **Ordinary Least Squares (OLS)**.  
In matrix form, the OLS estimator is:

$$
\hat{\Phi} = (X'X)^{-1} X'Y.
$$

This gives us a **unique estimate** of the VAR coefficients because OLS is well-defined and does not require any structural assumptions.

---

### Interpretation of $\hat{\Phi}$

- The matrix $\hat{\Phi}$ contains all estimated **lagged responses** of every variable to every other variable.
- It does **not** yet tell us how the system responds to **structural economic shocks** (e.g., “a monetary policy shock”).
- The errors $u_t$ still contain **contemporaneous correlation**:

$$
\Omega = E[u_t u_t'] \neq \text{diagonal}.
$$

This means the shocks in the reduced-form VAR are **not economically interpretable**.

To obtain **structural shocks** (orthogonal shocks), we must **identify** the model.

---




````python
import numpy as np
import pandas as pd

# ================================
# 1) PREPARE DATA
# ================================

# Use df_final directly (already cleaned with dropna)
Y = df_original[["2y Treas", "IP", "CPI", "EBP"]].values

T, K = Y.shape     # T = number of obs, K = 4 variables
p = 12             # VAR lag order
m=4               # number of variables

# ================================
# 2) BUILDING THE LAGGED MATRIX X
# ================================

X = []

# Each row of X contains [alpha, Y_{t-1}, ..., Y_{t-12}]
for i in range(p, T):
    lags = []
    for lag in range(1, p+1):
        lags.append(Y[i - lag])
    X.append(np.concatenate(lags))

X = np.array(X)

# Y_target = values for y_t corresponding to the lag structure
Y_target = Y[p:]

# Add intercept column
intercept = np.ones((X.shape[0], 1))
X = np.hstack((intercept, X))

print("Y_target shape:", Y_target.shape)  # (T-p, 4)
print("X shape:", X.shape)               # (T-p, 1 + 12*4)

# ================================
# 3) CHECK INVERTIBILITY OF X'X
# ================================

print("Rank of X:", np.linalg.matrix_rank(X))
print("Columns:", X.shape[1])

XtX = X.T @ X
print("Rank of X'X:", np.linalg.matrix_rank(XtX))
print("Shape of X'X:", XtX.shape)

print("Is X'X invertible?", np.linalg.matrix_rank(XtX) == XtX.shape[0])

````

````python
# ================================
# OLS ESTIMATION: PHI = (X'X)^(-1) X'Y
# ================================

# Compute X'X and X'Y
XtX = X.T @ X
XtY = X.T @ Y_target

# Try classical inversion
try:
    XtX_inv = np.linalg.inv(XtX)
    print("Successfully inverted X'X.")
except np.linalg.LinAlgError:
    print("X'X is not invertible. Cannot use the classical formula.")
    raise

# OLS estimator
PHI = XtX_inv @ XtY

print("\nPHI estimated with classical OLS:")
print(PHI)
print("PHI shape:", PHI.shape)   # should be (49, 4) in VAR(12) with 4 variables
````

In our case, the model has:

$$

$$

---

## Interpretation of the rows in $ \hat{\Phi} $

Each row of $ \hat{\Phi} $ corresponds to one **lagged variable** across all 9 equations of the system.


Each group of 4 rows in $ \hat{\Phi} $ corresponds to the coefficients associated with the 4 lagged variables for a given lag.

Row 0 → constant term

Rows 1–4 → coefficients for the 4 lagged variables at lag 1

Rows 5–8 → coefficients for the 4 lagged variables at lag 2

Rows 9–12 → coefficients for the 4 lagged variables at lag 3

…

Rows 45–48 → coefficients for teh 4 lagged variables at lag 12

Hence, we can think of $ \hat{\Phi} $ as containing the **columns** of each $ A_j $ matrix stacked vertically.

A_1' \\
A_2' \\
\vdots \\
A_{12}'
\end{bmatrix},
$$

where each $A_j'$ contributes **4 rows**, corresponding to its **4 columns**.

---

\begin{aligned}

A_2    &= \left( \hat{\Phi}[5:9, :] \right)' \\

Each group of 4 rows in $\hat{\Phi}$ corresponds to the coefficients associated with the 4 lagged variables at that specific lag.

each block must be transposed to reconstruct the proper $ A_j $ matrix in the VAR representation.

## Vector Moving Average (VMA) Representation

Consider a $p$-order **Vector AutoRegressive (VAR)** model in reduced form:

$$
y_t = \sum_{i=1}^{p} A_i y_{t-i} + u_t
$$

where $y_t$ is an $m \times 1$ vector of endogenous variables, $A_i$ are $m \times m$ coefficient matrices, and $u_t$ are reduced-form residuals with covariance matrix $\Omega = E(u_t u_t')$.

---

### Step 1. Writing the VAR in lag operator form

Using the **lag operator** $L$, where $L y_t = y_{t-1}$, we can rewrite the model as:

$$
(I_m - A_1 L - A_2 L^2 - \dots - A_p L^p) y_t = u_t
$$

Let’s define the **matrix lag polynomial**:

$$
A(L) = I_m - A_1 L - A_2 L^2 - \dots - A_p L^p
$$

Thus, the VAR can be expressed compactly as:

$$
A(L) y_t = u_t
$$

---

### Step 2. Inverting the VAR into the VMA form

If the process is **stationary**, the matrix polynomial $A(L)$ is **invertible**, and we can write:

$$
y_t = A(L)^{-1} u_t = \Phi(L) u_t
$$

where $\Phi(L)$ is an infinite-order matrix lag polynomial:

$$
\Phi(L) = I_m + \Phi_1 L + \Phi_2 L^2 + \Phi_3 L^3 + \dots = \sum_{j=0}^{\infty} \Phi_j L^j
$$

Therefore, the VAR can be rewritten as a **Vector Moving Average (VMA)** process:

$$
y_t = \sum_{j=0}^{\infty} \Phi_j u_{t-j}, \quad \Phi_0 = I_m
$$

This representation expresses the current value of $y_t$ as a (possibly infinite) weighted sum of current and past reduced-form shocks $u_t, u_{t-1}, u_{t-2}, \dots$.

---

### Step 3. Iterative derivation (intuition)

To understand the structure, consider a simple VAR(1):

$$
y_t = A_1 y_{t-1} + u_t
$$

Iterating forward, we get:

$$
\begin{aligned}
y_1 &= A_1 y_0 + u_1 \\
y_2 &= A_1 y_1 + u_2 = A_1(A_1 y_0 + u_1) + u_2 = A_1^2 y_0 + A_1 u_1 + u_2
\end{aligned}
$$

Assuming $y_0 = 0$, we obtain:

$$
y_2 = A_1 u_1 + u_2 = \sum_{j=0}^{1} A_1^j u_{2-j}
$$

and, more generally:

$$
y_t = \sum_{j=0}^{t-1} A_1^j u_{t-j}, \quad \text{with } A_1^0 = I_m
$$

This shows that for a VAR(1), each $y_t$ depends on **current and past shocks** weighted by powers of $A_1$.

---

### Step 4. Stationarity and invertibility condition

For a general $p$-order VAR, the **invertibility** of $A(L)$ (and thus the existence of a stable VMA representation) requires that the determinant equation:

$$
\big| I_m - A_1 z - A_2 z^2 - \dots - A_p z^p \big| = 0
$$

has all its roots **outside the unit circle**.  
This condition guarantees that the infinite sum $\sum_{j=0}^{\infty} \Phi_j L^j$ converges.

When this holds, the VAR is **stationary**, and the VMA representation exists.  
Hence, each variable can be expressed as a **stable linear combination** of past shocks.

---

### Step 5. From VMA to Structural VMA (SVMA)

If the structural form of the model is:

$$
A_0 y_t = \sum_{i=1}^p A_i^* y_{t-i} + \varepsilon_t, \quad E(\varepsilon_t \varepsilon_t') = \Sigma
$$

and the reduced-form errors are related by:

$$
u_t = A_0^{-1} \varepsilon_t,
$$

then substituting into the VMA gives the **Structural VMA (SVMA)** representation:

$$
y_t = \sum_{j=0}^{\infty} \Phi_j A_0^{-1} \varepsilon_{t-j}, \qquad \Phi_0 = I_m
$$

This representation is crucial for identifying the **dynamic propagation of structural shocks** $\varepsilon_t$ through time.

---

### Step 6. Connection to Impulse Response Functions (IRFs)

Once the SVMA is obtained, we can directly compute the **Impulse Response Functions (IRFs)**.

An IRF traces the effect of a **unit structural shock** to one variable on all other variables in the system over time.  
From the SVMA representation:

$$
y_{t+h} = \sum_{j=0}^{\infty} \Phi_j A_0^{-1} \varepsilon_{t+h-j}
$$

the **Impulse Response at horizon $h$** is given by:

$$
IRF(h) = \Phi_h A_0^{-1}
$$

Each matrix $\Phi_h A_0^{-1}$ describes how a one-time innovation in $\varepsilon_t$ affects the system in period $t+h$.

Thus, the VMA (and its structural counterpart SVMA) provides the **foundation for dynamic analysis** — it allows us to interpret the estimated VAR coefficients in terms of **economic shocks** and their **propagation mechanisms**.

---




# The Companion Form System

The **companion form** is essentially a reformulation of the VAR($p$) model as a **VAR(1)** system in which:

- The **first block equation** reproduces the original VAR dynamics.
- The **remaining block equations** serve to **shift the lagged variables forward**, using identities.

---

Starting from the $p$-order VAR:

$$
y_t = A_1 y_{t-1} + A_2 y_{t-2} + \dots + A_p y_{t-p} + u_t
$$

we define the *stacked vector* of all current and lagged endogenous variables:

$$
Y_t =
\begin{bmatrix}
y_t \\
y_{t-1} \\
y_{t-2} \\
\vdots \\
y_{t-(p-1)}
\end{bmatrix}
$$

Then the system $Y_t = A Y_{t-1} + U_t$ can be interpreted equation by equation as:

$$
\begin{cases}
y_t = A_1 y_{t-1} + A_2 y_{t-2} + \dots + A_p y_{t-p} + u_t, & \text{(the original VAR equation)} \\[6pt]
y_{t-1} = I_m y_{t-1}, & \text{(identity: shifts lag 1)} \\[6pt]
y_{t-2} = I_m y_{t-2}, & \text{(identity: shifts lag 2)} \\[6pt]
\vdots \\[4pt]
y_{t-(p-1)} = I_m y_{t-(p-1)}, & \text{(identity: shifts lag p–1)}
\end{cases}
$$

or, equivalently in matrix notation:

$$
\underbrace{
\begin{bmatrix}
y_t \\ y_{t-1} \\ y_{t-2} \\ \vdots \\ y_{t-(p-1)}
\end{bmatrix}}_{Y_t}
=
\underbrace{
\begin{bmatrix}
A_1 & A_2 & \dots & A_p \\
I_m & 0 & \dots & 0 \\
0 & I_m & \dots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \dots & I_m &
\end{bmatrix}}_{A}
\underbrace{
\begin{bmatrix}
y_{t-1} \\ y_{t-2} \\ y_{t-3} \\ \vdots \\ y_{t-p}
\end{bmatrix}}_{Y_{t-1}}
+
\underbrace{
\begin{bmatrix}
u_t \\ 0 \\ 0 \\ \vdots \\ 0
\end{bmatrix}}_{U_t}
$$


For a VAR(1) in the VMA representation obtained directly from the estimated VAR coefficients, i.e. $ y_t = \sum_{j=0}^{t-1}A^j_1 u_{t-j} $, a convenient result of the companion form notation is:

$ Y_t = \sum_{j=0}^{t-1} A^j U_{t-j}, \quad A^0 = I_{mp} $


If stationarity/invertibility is satisfied, $ A^j $ goes to zero as $ j $ increases to infinity, i.e. the effects of past shocks tend to vanish over time. 
In our case we used variables in levels so even if the inveribility is not satisfied thanks to the Companion Form we can still invert the process and obtain the VMA representation. As we will see later some variables' IRF will not return to zero since they are non-stationary and we get the response in levels (cumulative response).

## Dimension of Companion Matrix of a VAR(12) with 4 Variables


- **Top block:** $4 \times 48$  
- **Bottom block:** $44 \times 48$  
- **Full companion matrix:** $48 \times 48$

Let’s explain why this is the case.


We have a VAR(12) with 4 variables

$m = 4 \quad$ (number of variables) 
 
$p = 12 \quad$ (number of lags)

The state vector of the equivalent VAR(1) system is:

$X_t =
\begin{bmatrix}
y_t' & y_{t-1}' & y_{t-2}' & \cdots & y_{t-11}'
\end{bmatrix}'$

where each $y_t$ is a $4 \times 1$ vector.

Hence:

$\dim(X_t) = 4 \times 12 = 48$

Therefore, the **companion matrix must be $48 \times 48$**.

---

### Structure of the Companion Matrix

The standard form of the companion matrix is:

$A_{\text{comp}} =
\begin{bmatrix}
A_1 & A_2 & \cdots & A_{12} \\
I_4 & 0 & \cdots & 0 \\
0 & I_4 & \cdots & 0 \\
\vdots & & \ddots & \vdots \\
0 & 0 & \cdots & I_4
\end{bmatrix}$

**Dimensions:**

- The **top block row** contains $A_1, \dots, A_{12}$, each of size $4 \times 4$:

  $4 \times (4 \times 12) = 4 \times 48$

- The **bottom block** contains 11 identity matrices $I_4$:

  $(11 \times 4) \times (11 \times 4 + 4) = 44 \times 48$

So the entire companion matrix is indeed:

$A_{\text{comp}} \in \mathbb{R}^{48 \times 48}$

---

### The Constant Does *Not* Enter the Companion Matrix

The constant (intercept) does **not** belong to the linear dynamics of the VAR:

$y_t = c + A_1 y_{t-1} + A_2 y_{t-2} + \dots + A_{12} y_{t-12} + u_t$

The companion matrix only captures the **dynamic part**:

$X_t = A_{\text{comp}} X_{t-1} + B u_t + (\text{constant term})$

Thus, the **companion matrix** includes only the lag coefficients $A_1, \dots, A_{12}$.  
The constant vector $c$ is handled separately as an additive term.

To include the intercept, you can define an **augmented constant vector**:

$d_{\text{comp}} =
\begin{bmatrix}
c \\
0 \\
\vdots \\
0
\end{bmatrix}$

But this does **not** change the size of $A_{\text{comp}}$.

## 5.2 Identification: From Reduced-Form VAR to Structural VAR

Once we estimate the reduced-form VAR and obtain the residuals $u_t$ and their covariance matrix 
$\Omega = E[u_t u_t']$, we need to uncover the underlying **structural shocks** that truly drive the economy.  
The reduced-form errors $u_t$ can be viewed as linear combinations of these structural shocks, denoted by $\varepsilon_t$.

Recall the structural and reduced-form representations:

$$
A_0 Y_t = \sum_{i=1}^{p} A_i^* Y_{t-i} + \varepsilon_t \quad \text{(SVAR)}
$$

$$
Y_t = \sum_{i=1}^{p} A_i Y_{t-i} + u_t \quad \text{(Reduced-Form VAR)}
$$

The identification problem consists in finding a unique mapping between **reduced-form residuals** $u_t$ and **structural shocks** $\varepsilon_t$:


---

### Relationship between Reduced-Form and Structural Errors

We assume that the economy is driven by a set of **serially uncorrelated structural shocks** $\varepsilon_t$, with:

$$
E[\varepsilon_t] = 0, \quad \text{and} \quad Var(\varepsilon_t) = \Sigma_\varepsilon = I.
$$

This means the structural shocks are **mutually uncorrelated** and have **unit variance**.  
The link between reduced-form and structural errors is linear:

$$
u_t = B \varepsilon_t
$$

where $B$ is a $(4 \times 4)$ **impact matrix** that maps each structural shock into the reduced-form innovations.

---

### Covariance Relationship

Taking variances on both sides of $u_t = B \varepsilon_t$, we obtain the key covariance identity:

$$
\Omega = E[u_t u_t'] = E[B \varepsilon_t \varepsilon_t' B'] = B E[\varepsilon_t \varepsilon_t'] B' = B \Sigma_\varepsilon B'.
$$

Since we assume $\Sigma_\varepsilon = I$, this simplifies to:

$$
\boxed{\Omega = B B'}
$$

Thus, $\Omega$ (which we estimated directly from the reduced-form VAR residuals) can be factored as the product of $B$ and its transpose.  

---

### Identification Problem

The challenge is that $\Omega = B B'$ does **not uniquely determine** the matrix $B$:
- There are infinitely many matrices $B$ that satisfy this equation.
- Even if we know $\Omega$ perfectly, we can always find many different $B$ that yield the same $\Omega$.

This is known as the **identification problem** in Structural VARs.

To identify $B$ (and therefore the structural shocks $\varepsilon_t$), the econometrician must impose additional **restrictions**—either theoretical (based on economic reasoning) or empirical (such as proxy instruments or sign restrictions).

---

## 5.2.1 Proxy-VAR (IV-VAR) Identification

To solve the **identification problem** of the Structural VAR (SVAR) and the Reduced-Form VAR (RF-VAR), 
different strategies can be adopted.  

### Classical Identification Methods

1. **Exclusion Restrictions** (Cholesky / Recursive Identification)  
   → assumes a lower-triangular structure of the impact matrix $B$.

2. **AB-Model** (Blanchard and Perotti, 2002)  
   → imposes both short-run (A) and contemporaneous (B) restrictions.

3. **Long-Run (LR) Restrictions** (e.g., Blanchard and Quah, 1989).

4. **Sign Restrictions**  
   → identification based on sign patterns of impulse responses.

5. **Proxy-VAR (IV-VAR)**  
   → used when we do **not believe** in recursive or set-based restrictions.  
     This approach uses an **external variable (proxy)** to identify a specific structural shock.

---

If we impose **Cholesky (recursive) restrictions**, we assume that $B$ is **lower-triangular**, i.e.:

$$
B = 
\begin{bmatrix}
\star & 0 & 0 & 0 \\
\star & \star & 0 & 0 \\
\star & \star & \star & 0 \\
\star & \star & \star & \star
\end{bmatrix}.
$$

However, this assumption is often **hard to justify economically**.  
In the Proxy-VAR framework, instead, we **do not need to identify the full matrix $B$** —  
we only aim to recover the **first column $b_1$**, which captures the contemporaneous impact of the structural shock of interest (e.g., the monetary policy shock).

---

### What is a Proxy-VAR?

The Proxy-VAR (or IV-VAR) allows the identification of specific **structural shocks** in a VAR system **without imposing strong restrictions** on contemporaneous interactions or long-run relations.  

Essentially, the innovation of interest (e.g., the **monetary policy shock**) is identified by using an **external variable** that is:
- correlated with the **shock of interest**, and  
- uncorrelated with all other shocks in the VAR.  

Under these assumptions, the external variable acts as a **valid instrument** for the structural shock — hence the name **Instrumental Variable VAR (IV-VAR)** or **Proxy-VAR**.


---

### Decomposition of the Impact Matrix

We can decompose the impact matrix $B$ as:

$$
B = [\, b_1 \; B_2 \,],
$$

where:
- $b_1$ is the $m \times 1$ vector that describes the **impact of the structural shock of interest** $\varepsilon_{1t}$,  
- $B_2$ is the $(m \times (m-1))$ matrix of the remaining columns, associated with the **other structural shocks** $\varepsilon_{2t}$.

Let $\varepsilon_t$ be partitioned conformably as:

$$
\varepsilon_t =
\begin{bmatrix}
\varepsilon_{1t} \\
\varepsilon_{2t}
\end{bmatrix},
$$

with:
- $\varepsilon_{1t}$ → scalar ($1 \times 1$), the **structural shock of interest**,  
- $\varepsilon_{2t}$ → vector ($(m-1) \times 1$), the **remaining structural shocks**.

Then the relation $u_t = B \varepsilon_t$ can be written explicitly as:

$$
u_t =
[\, b_1 \; B_2 \,]
\begin{bmatrix}
\varepsilon_{1t} \\
\varepsilon_{2t}
\end{bmatrix}
=
b_1 \varepsilon_{1t} + B_2 \varepsilon_{2t}.
$$
Our goal is to **identify $b_1$**, which measures how the structural shock of interest contemporaneously affects all variables in $Y_t$.

---

### Matrix Dimensions and Decomposition

The matrix $B$ is square, of dimension $m \times m$:

$$
B =
\begin{bmatrix}
| & | &  & | \\
b_1 & b_2 & \dots & b_m \\
| & | &  & |
\end{bmatrix},
\quad B \in \mathbb{R}^{m \times m}.
$$



### Example: Our Case ($m = 4$)

In our VAR with four variables ($m = 4$), the impact matrix $B$ can be represented as:

$$
B =
\begin{bmatrix}
| & | & | & | \\
b_{11} & b_{12} & b_{13} & b_{14} \\
b_{21} & b_{22} & b_{23} & b_{24} \\
b_{31} & b_{32} & b_{33} & b_{34} \\
b_{41} & b_{42} & b_{43} & b_{44} \\
| & | & | & |
\end{bmatrix}
=
\begin{bmatrix}
b_1 & B_2
\end{bmatrix}.
$$

Each column of $B$ corresponds to the **contemporaneous effect** of a specific structural shock on all the $m$ variables in $Y_t$.  

- $b_1$ → column vector ($m \times 1$) associated with the **shock of interest** (e.g., the monetary policy shock).  
- $b_2, \dots, b_m$ → columns ($m \times (m-1)$) associated with **other structural shocks** (e.g., supply, demand, financial shocks, etc.).

---

We can rewrite the structural relationship as:

$$
u_t = \begin{bmatrix}
| & | & | & | \\
b_{11} & b_{12} & b_{13} & b_{14} \\
b_{21} & b_{22} & b_{23} & b_{24} \\
b_{31} & b_{32} & b_{33} & b_{34} \\
b_{41} & b_{42} & b_{43} & b_{44} \\
| & | & | & |
\end{bmatrix}
\begin{bmatrix}
\varepsilon_t^{mp} \\
\varepsilon_t^{(2)} \\
\varepsilon_t^{(3)} \\
\varepsilon_t^{(4)}
\end{bmatrix}
$$

Basically $b_1$ teel ls us how a one-unit innovation in the monetary policy shock $\varepsilon_t^{mp}$ affects each of the four variables in $Y_t$ contemporaneously.

---


### Dimensions Summary

| Symbol | Description | Dimension |
|:--------|:-------------|:-----------|
| $u_t$ | Reduced-form residuals | $(m \times 1)$ |
| $B$ | Impact (mapping) matrix | $(m \times m)$ |
| $b_1$ | Column of $B$ for the structural shock of interest | $(m \times 1)$ |
| $B_2$ | Remaining columns of $B$ | $(m \times (m-1))$ |
| $\varepsilon_{1t}$ | Structural shock of interest | $(1 \times 1)$ |
| $\varepsilon_{2t}$ | Remaining structural shocks | $((m-1) \times 1)$ |

---

### Assumptions

1. **Stationarity:**  
   The VAR is stationary and of finite order $p$.

2. **Invertibility of $B$:**  
   $B$ must be invertible ($\det(B) \neq 0$), to allow recovery of structural shocks from reduced-form residuals:
   $$
   \varepsilon_t = B^{-1} u_t.
   $$

3. **Orthogonality of Structural Shocks:**  
   $$
   E[\varepsilon_t \varepsilon_t'] = 
   \Sigma_\varepsilon =
   \begin{bmatrix}
   \sigma_1^2 & 0 \\
   0 & \Sigma_2
   \end{bmatrix}.
   $$
   This implies that the structural shock of interest $\varepsilon_{1t}$ is **uncorrelated** with the remaining $(m-1)$ shocks in $\varepsilon_t$.

---

### Proxy Identification Strategy

Since structural shocks $\varepsilon_t$ are not observable, we introduce an **external proxy variable** $z_t$ that behaves like the structural shock $\varepsilon_{1t}$.

#### The proxy $z_t$ is valid if:

1. **Relevance:**  
   $z_t$ is correlated with the structural shock of interest:
   $$
   E[\varepsilon_{1t}(z_t - \bar{z})] = \phi \neq 0.
   $$

2. **Exogeneity:**  
   $z_t$ is uncorrelated with all the other structural shocks:
   $$
   E[\varepsilon_{2t}(z_t - \bar{z})] = 0.
   $$

Under these assumptions, $z_t$ can serve as a **valid instrument** for $\varepsilon_{1t}$.

---

### Deriving the Proxy-VAR Fundamental Relationship

Starting from:

$$
u_t = b_1 \varepsilon_{1t} + B_2 \varepsilon_{2t},
$$

multiply both sides by $(z_t - \bar{z})$ and take expectations:

$$
E[u_t (z_t - \bar{z})] = E[b_1 \varepsilon_{1t} (z_t - \bar{z})] + E[B_2 \varepsilon_{2t} (z_t - \bar{z})].
$$

By the **exogeneity assumption**:

$$
E[\varepsilon_{2t} (z_t - \bar{z})] = 0,
$$

and by **relevance**:

$$
E[\varepsilon_{1t} (z_t - \bar{z})] = \phi \neq 0.
$$

Therefore:

$$
\boxed{E[u_t (z_t - \bar{z})] = \phi b_1}
\quad \Rightarrow \quad
b_1 \propto E[u_t (z_t - \bar{z})].
$$

This is the **fundamental equation of the Proxy-VAR**.  
It links the covariance between the external instrument $z_t$ and the reduced-form residuals $u_t$ to the **impact vector $b_1$**, up to scale $\phi$.

---


### Construction and Identification of the Instrument $z_t$

To identify the structural monetary policy shock within the Proxy-VAR framework,  
we use **high-frequency monetary policy surprises** around FOMC announcements as external instruments.  
This identification strategy follows the approach of **Gertler and Karadi (2015)**,  
and is now standard in the **high-frequency identification literature**.

---

### High-Frequency Identification

To identify this impact vector $b_1$, we use **high-frequency surprises** in interest rates  
recorded around FOMC announcements. These are obtained from financial markets,  
for example changes in **Eurodollar futures** (ED1–ED4) in narrow time windows  
(typically 30 minutes before and after the announcement).

Let $z_t$ denote this external instrument.  
It summarizes the **unexpected changes** (or *surprises*) in monetary policy decisions within each month.  
In practice, since the high-frequency surprises are available at a daily or intraday frequency,  
we aggregate them to the monthly level by summing across all events within each month.

---

### Validity Conditions for the Instrument

For $z_t$ to be a valid instrument for the **monetary policy shock** $\varepsilon_{1t}$,  
two conditions must hold — these are the **relevance** and **exogeneity** assumptions.

#### (a) Relevance Condition

The instrument must be correlated with the structural shock of interest:

$$
E[z_t \varepsilon_{1t}] = \phi \neq 0.
$$

This ensures that $z_t$ carries information about the monetary policy shock.  
Intuitively, $z_t$ must move whenever the Fed surprises the market.

#### (b) Exogeneity Condition

The instrument must be uncorrelated with all **other structural shocks** in the system:

$$
E[z_t \varepsilon_{2t}] = 0,
$$

where $\varepsilon_{2t}$ collects all shocks *other than* monetary policy (e.g., demand, supply, or financial shocks).

This assumption guarantees that $z_t$ identifies only the monetary policy component of variation in $u_t$.

---

### The Endogeneity Problem

However, in practice, these assumptions may fail — particularly the **exogeneity condition**.  
As shown in the analysis of Section 3, **high-frequency monetary policy surprises ($z_t$)** can be *predictable*  
using macroeconomic and financial variables known before the FOMC announcement.  

If $z_t$ is correlated with such pre-announcement information,  
then it will also be correlated with non-policy shocks that move with macroeconomic conditions.  
This introduces **endogeneity**: $z_t$ may capture not only monetary shocks,  
but also systematic responses of the Fed to the economic outlook.

In that case, $E[z_t \varepsilon_{2t}] \neq 0$,  
and $z_t$ fails to satisfy the exogeneity condition.

---

### Orthogonalization of the Instrument

To address this issue, we follow **Bauer and Swanson (2021)** and construct an **orthogonalized instrument**,  
denoted by $z_t^{\perp}$, which removes the predictable (endogenous) component of $z_t$.

The orthogonalized instrument is obtained by regressing $z_t$ on the set of macro-financial predictors  
known before the announcement (those in Table 1 of the paper, e.g., employment growth, yield curve slope,  
stock market returns, commodity prices, etc.) and taking the residuals:

$$
z_t^{\perp} = z_t - \hat{X}_{t-}'\hat{\beta},
$$

where $X_{t-}$ contains pre-announcement macroeconomic variables.  
The residual $z_t^{\perp}$ represents the part of the monetary policy surprise  
that cannot be explained by previous macroeconomic conditions.

---

### Why the Orthogonalization Works

The orthogonalized surprise $z_t^{\perp}$ is expected to:
- **Satisfy exogeneity**, because it removes correlations between $z_t$ and other structural shocks $\varepsilon_{2t}$:
  $$
  E[z_t^{\perp} \varepsilon_{2t}] = 0,
  $$
- **Retain relevance**, since it preserves the unexpected component of the monetary policy shock:
  $$
  E[z_t^{\perp} \varepsilon_{1t}] = \phi \neq 0.
  $$

Therefore, $z_t^{\perp}$ is a **cleaner instrument** that isolates the true monetary policy innovation.

---

### Intuitive Interpretation

- The **raw instrument** $z_t$ captures monetary policy surprises,  
  but may still reflect the Fed’s *systematic response* to news about the economy.

- The **orthogonalized instrument** $z_t^{\perp}$ removes that systematic component,  
  leaving only the *unexpected part* of monetary policy behavior.

In this sense, $z_t^{\perp}$ better satisfies the assumptions required for **causal identification** in the Proxy-VAR.

---

### Summary of the Instrument Construction Process

| Step | Description | Formula |
|------|--------------|----------|
| 1 | Obtain high-frequency monetary policy surprises | $z_t$ from Eurodollar futures (ED1–ED4) |
| 2 | Aggregate to monthly frequency | $\displaystyle z_t = \sum_{\text{events in month } t} \text{MPS}_i$ |
| 3 | Orthogonalize against pre-announcement information | $z_t^{\perp} = z_t - \hat{X}_{t-}'\hat{\beta}$ |
| 4 | Use $z_t^{\perp}$ as instrument in Proxy-VAR | $E[u_t (z_t^{\perp} - \bar{z}^{\perp})] = \phi b_1$ |

---

Thus, the Proxy-VAR exploits the covariance between the reduced-form residuals $u_t$  
and the *orthogonalized instrument* $(z_t^{\perp} - \bar{z}^{\perp})$ to identify the **impact vector** $b_1$,  
which measures the contemporaneous effects of a structural monetary policy shock on all variables in the VAR.

$$
\boxed{E[u_t (z_t^{\perp} - \bar{z}^{\perp})] = \phi b_1}
$$

This moment condition is the **empirical cornerstone** of the Proxy-VAR identification strategy.

---


### Sample Alignment Between the VAR and the Instrument

In proxy-SVAR identification, it is common that the vector autoregression (VAR) is 
estimated over a long historical sample, while the external instrument (the proxy variable) 
is available only for a shorter subsample. This situation does **not** pose a problem for 
estimation or identification.

Consider the reduced-form VAR:

$$
y_t = C + A_1 y_{t-1} + \cdots + A_p y_{t-p} + u_t,
$$

estimated over the full sample period (e.g., 1973–2020). From this estimation we obtain
the reduced-form residuals $u_t$, which exist for the entire VAR sample.  
The external instrument $z_t$, however, may only be available for a later subperiod
(e.g., starting in 1988). In this case, identification is conducted **only over the period in
which both the reduced-form shocks and the instrument are jointly observed**.

This is standard in the proxy-SVAR literature. Because the instrument is required only for
the structural identification step—not for the reduced-form estimation—the VAR may be 
estimated on a longer sample than the instrument. The identification is then performed on 
the overlapping subsample:

$$
\{t : z_t \text{ is observed} \} \subset \{t : u_t \text{ is observed} \}.
$$

## Identification of $b_1$: From the Fundamental Formula to the Normalized Impact Vector

We start from the **Fundamental Formula of the Proxy-VAR**, which links the reduced-form residuals of the VAR to the instrument (proxy variable):

$$
E[u_t (z_t - \bar{z})] = b_1 \phi
$$

This relationship is **central** to the Proxy-VAR identification strategy, as it allows us to recover the structural impact vector $b_1$ up to a scaling factor $\phi$.

---

### 1. The Identification Problem

The equation above can be interpreted as follows:

- On the **left-hand side**, $E[u_t (z_t - \bar{z})]$ is *observable*, because it depends on the estimated reduced-form residuals $u_t$ and the demeaned instrument $(z_t - \bar{z})$.
- On the **right-hand side**, we have $b_1 \phi$.  
  Here, $b_1$ is the **impact vector** of the structural shock we want to identify, while $\phi$ is a **scaling factor** that we cannot directly observe.

Hence, the main difficulty is that $b_1$ and $\phi$ are confounded:  
we can only observe their product $b_1 \phi$.

---

### 2. Introducing the Extended Vector $v$

Define:

$$
v = E[u_t (z_t - \bar{z})]
$$

Then, by the fundamental formula:

$$
v = b_1 \phi
$$

Since $\phi$ is a scalar, $v$ and $b_1$ are *collinear*.  
Therefore, $v$ points in the same direction as $b_1$, but it has a different magnitude.

---

### 3. The GMM (Moment) Representation

#### From $v = b_1 \phi$ to the Proxy-VAR Moment Condition

We start from the vector equation:

$$
v = E[u_t (z_t - \bar{z})] = b_1 \phi
$$

where:
- $v$ is $(m \times 1)$,  
- $b_1$ is $(m \times 1)$,  
- $\phi$ is a scalar.

Our goal is to obtain a **scalar identifying condition** that allows us to identify $\phi$.  
Since $\phi$ is a scalar we must transform this vector equation into a **scalar form**. This will imply to use a **quadratic form**.

---

#### **Step 1 — Pre-multiply by $E[(z_t - \bar{z})u_t']$**

We pre-multiply the vector equation by the transpose of the same expectation.  
This operation turns the $(m \times 1)$ vector into a scalar:

$$
E[(z_t - \bar{z})u_t'] \, v
$$

**Dimensions:**

$$
(1 \times m) \times (m \times 1) = (1 \times 1)
$$

➡️ Now we have a **scalar**, the first step toward isolating $\phi$.

---

#### **Step 2 — Insert the weighting matrix $\Omega^{-1}$**

Since the reduced-form residuals $u_t$ are correlated across equations,  
we need to "whiten" them by inserting the inverse covariance matrix $\Omega^{-1}$  
between the two expectation terms. This corrects for cross-equation correlations:

$$
E[(z_t - \bar{z})u_t'] \, \Omega^{-1} \, v
$$

**Dimensions:**

$$
(1 \times m) \times (m \times m) \times (m \times 1) = (1 \times 1)
$$

➡️ The result remains a scalar, but now properly weighted by $\Omega^{-1}$.

---

#### **Step 3 — Substitute $v = b_1 \phi$ and post-multiply**

Finally, we replace $v$ with $b_1 \phi$ and again multiply by its transpose term.  
This step creates the desired **scalar quadratic form** in $b_1$ and $\phi$:

$$
E[(z_t - \bar{z})u_t'] \, \Omega^{-1} \, E[u_t (z_t - \bar{z})]
= (\phi b_1') \, \Omega^{-1} \, (\phi b_1)
= \phi^2 b_1' \Omega^{-1} b_1
$$

**Dimensions:**

$$
(1 \times m) \times (m \times m) \times (m \times 1) = (1 \times 1)
$$

➡️ Both sides are now **scalars**, and the right-hand side takes the form of a **quadratic form**,  
which is key for identifying $\phi^2$ and normalizing $b_1$.

---

✅ **Final Result:**

Recall that the reduced-form errors and the structural shocks are related by:

$$
u_t = B \varepsilon_t
$$

so that the variance-covariance matrix of reduced-form errors is:

$$
\Omega = E[u_t u_t'] = B \Sigma_\varepsilon B'
$$

Assuming $\Sigma_\varepsilon = I$, we have $\Omega = B B'$.



$$
E[(z_t - \bar{z})u_t'] \, \Omega^{-1} \, E[u_t (z_t - \bar{z})]
=\phi^2 b_1' \Omega^{-1} b_1
$$

That is,
$$ E[(z_t - \bar{z}) u_t'] E[u_t u_t']^{-1} E[u_t (z_t - \bar{z})] = \phi^2 b_1' (B \Sigma_\varepsilon B')^{-1} b_1$$



This is the **Proxy-VAR moment condition**,  
a scalar equality linking observable covariances to the structural parameters $(b_1, \phi^2)$.  
It provides the basis for the later normalization step and the identification of the impact vector $b_1$.

---


This is a **quadratic form**, which leads us to the GMM-like expression for $\phi^2$:

$$
\phi^2 = \sigma_{z,1}^2 E[(z_t - \bar{z}) u_t'] E[u_t u_t']^{-1} E[u_t (z_t - \bar{z})]
$$

However, since $\varepsilon_{1,t}$ (the structural shock of interest) is **unobserved**,  
its variance $\sigma_{z,1}^2 = \text{Var}(\varepsilon_{1,t})$ is **not known**.

---

### 4. Normalization

Following standard practice (see Mertens & Ravn, 2013),  
we can choose an **arbitrary variance normalization** for the structural shock:

$$
E[\varepsilon_{1,t}^2] = \sigma_{z,1}^2 = 1
$$

This implies we are working with a **unit-variance shock**, which does not affect the impulse response functions — it only fixes the scaling.

Given this normalization, we obtain the closed-form solution for $b_1$:

$$
b_1 = \pm \frac{E[u_t (z_t - \bar{z})]}{\left( E[(z_t - \bar{z}) u_t'] E[u_t u_t']^{-1} E[u_t (z_t - \bar{z})] \right)^{1/2}}
$$

This is the **normalized impact vector** associated with the structural shock identified by the instrument $z_t$.

---


## Practical Implementation of the Proxy-VAR Identification

### 1. Setting up the data

We start from two main inputs:

- The **reduced-form residuals** from the VAR, $u_t$, stacked as a $T \times n$ matrix (with $n = 4$ in our case: 2Y Treasury yield, Industrial Production, CPI, and Excess Bond Premium).
- The **proxy instrument**, $z_t$, which corresponds to the *Monetary Policy Surprises (MPS)* or their orthogonalized version (*MPS_ORTH*).

For each instrument, we estimate one structural shock, i.e. the **monetary policy shock**.

---

### 2. Computing the sample analogs of the theoretical moments

In theory, we have:

$$
v = E[u_t (z_t - \bar{z})]
$$

In practice, we approximate this expectation with the **sample covariance** between the residuals and the demeaned instrument:

$$
\hat{v} = \frac{1}{T} \sum_{t=1}^{T} u_t (z_t - \bar{z})
$$

This produces a $n \times 1$ vector $\hat{v}$, which is the empirical counterpart of $b_1 \phi$.

---

### 3. Computing the covariance of reduced-form residuals

The variance-covariance matrix of the reduced-form errors is:

$$
\hat{\Omega} = E[u_t u_t'] \approx \frac{1}{T} \sum_{t=1}^{T} u_t u_t'
$$

This matrix $\hat{\Omega}$ summarizes how the residuals of the different VAR equations are correlated contemporaneously.

---

From the theoretical formula for $b_1$ under unit-variance normalization, we have:

$$
b_1 = \frac{E[u_t (z_t - \bar{z})]}{\sqrt{E[(z_t - \bar{z}) u_t'] E[u_t u_t']^{-1} E[u_t (z_t - \bar{z})]}}
$$

The numerator corresponds to the covariance between the reduced-form residuals and the instrument (i.e. $\hat{v}$),  
and the denominator rescales this vector so that the identified structural shock has unit variance.

In code, this is implemented as:

```python
cov_uz = (u.T @ z) / T                  # → 𝑣̂ = E[u_t (z_t - z̄)]
Sigma_u = (u.T @ u) / T                 # → Ω̂ = E[u_t u_t′]
inv_Sigma = np.linalg.inv(Sigma_u)
denom = np.sqrt((cov_uz.T @ inv_Sigma @ cov_uz)[0,0])
```

$$
\hat{b}_1 = \frac{\hat{v}}{\sqrt{\hat{v}' \hat{\Omega}^{-1} \hat{v}}}
$$

This denominator plays the same role as dividing by $\phi$ —  
it rescales the vector so that the identified shock has **unit variance**.

### Geometric interpretation

- $\hat{v}$ is the **projection** of the instrument $z_t$ onto the space spanned by the residuals $u_t$.
- $b_1$ is the **direction** of that projection — i.e., the direction in which $z_t$ moves the system contemporaneously.
- Normalizing by $\|\hat{v}\|$ (or more precisely $\sqrt{\hat{v}' \hat{\Omega}^{-1} \hat{v}}$) makes the identified shock a **unit vector** in that direction.

So the code’s formula:

```python
b1 = cov_uz / denom




````python
import numpy as np

# --- Residual columns from your VAR ---
residual_cols = [
    "RESIDUALS 2Y_TREASURY",
    "RESIDUALS INDPRO",
    "RESIDUALS CPI",
    "RESIDUALS EBP",
    
]

# --- Extract residual matrix (T × n) ---
residuals = merged_clean[residual_cols].values

# --- Extract proxies ---
z = merged_clean["MPS"].values
z_orth = merged_clean["MPS_ORTH"].values



def proxy_svar_identification(residuals, proxy):
    """
    Identifies structural impact vector b1 using a proxy SVAR instrument.
    """
    proxy = proxy.reshape(-1, 1)

    T = min(len(proxy), len(residuals))
    u = residuals[:T]
    z = proxy[:T]

    # Cov(u, z)
    cov_uz = (u.T @ z) / T   # shape (n,1)

    # Covariance of residuals
    Sigma_u = (u.T @ u) / T
    inv_Sigma = np.linalg.inv(Sigma_u)

    # Normalization denominator
    denom = np.sqrt((cov_uz.T @ inv_Sigma @ cov_uz)[0,0])

    # Structural impact vector
    b1 = (cov_uz / denom).flatten()

    return b1


# --- Compute structural impact vectors ---
b1_z = proxy_svar_identification(residuals, z)
b1_z_orth = proxy_svar_identification(residuals, z_orth)

print("b1_z       =", b1_z)
print("b1_z_orth =", b1_z_orth)
````

## Computing Impulse Response Functions (IRFs) from the Identified $b_1$

After identifying the structural impact vector $b_1$ through the Proxy-VAR method,  
the next step is to trace the **dynamic propagation** of this structural shock over time.  
This is done by computing the **Impulse Response Functions (IRFs)**.

---

### The Structural Representation and the Role of $b_1$

The structural shocks $\varepsilon_t$ and the reduced-form residuals $u_t$ are related by:

$$
u_t = B \varepsilon_t,      
$$

where $B$ is the structural impact matrix.

In the Proxy-VAR, we do **not** identify the entire $B$ matrix,  
but only its **first column** $b_1$, associated with the identified structural shock —  
for example, the **monetary policy shock**.

This column vector $b_1$ captures the *contemporaneous impact* of a one-standard-deviation monetary policy shock on all variables in the VAR.

---

### Building the Structural Impact Matrix for the Identified Shock

Even if we only have $b_1$, that is sufficient to compute the IRFs for that particular shock.  
We construct a **restricted version** of the impact matrix, denoted $B_{comp}^{(1)}$, as follows:

$$
B_{comp}^{(1)} =
\begin{bmatrix}
b_1 \\ 0 \\ 0 \\ \vdots \\ 0
\end{bmatrix}.
$$

Here:
- the first $m$ rows contain the contemporaneous impact $b_1$,  
- and the remaining $(mp - m)$ rows are zeros,  
  because the shock affects the system only through the current variables, not the lagged ones.

This matrix allows the identified structural shock to enter the **companion system** dynamics.

In code, this is represented as:

```python
  # Initial state vector size: m*p
    state = np.zeros((m*p, 1))

    # Shock at time 0 enters in the first m positions (current variables)
    state[:m, 0] = b1  #This code create our B_comp matrix: first m elements are b1, rest are zeros
```

---

### Dynamic Propagation of the Structural Shock

Let $A_{comp}$ denote the **companion matrix** of dimension $(mp \times mp)$  
that contains all the estimated autoregressive dynamics from the reduced-form VAR.

$$
Y_t = A_{comp} Y_{t-1} + U_t,
$$

Since 

$$
U_t = B_{\text{comp}} \, \varepsilon_t
$$


$$
\begin{pmatrix}
u_t \\
0 \\
\vdots \\
0
\end{pmatrix}
=
\begin{pmatrix}
B \\
0 \\
\vdots \\
0
\end{pmatrix}
\varepsilon_t
$$


The evolution of the state vector $Y_t$ can be written as:

$$
Y_t = A_{comp} Y_{t-1} + B_{comp} \varepsilon_t,
$$

and when we focus only on the identified structural shock (the first one), we get:

$$
B_{\text{comp}} \neq B_{\text{comp}}^{(1)}
$$


$$
B_{\text{comp}} =
\begin{pmatrix}
B \\
0 \\
\vdots \\
0
\end{pmatrix}
\in \mathbb{R}^{mp \times m},
\qquad
B_{\text{comp}}^{(1)} =
\begin{pmatrix}
b_1 \\
0 \\
\vdots \\
0
\end{pmatrix}
\in \mathbb{R}^{mp \times 1}
$$


$$
Y_t = A_{comp} Y_{t-1} + B_{comp}^{(1)} \varepsilon_{1,t}.
$$

To study how the system reacts over time to a one-standard-deviation shock $\varepsilon_{1,t}$,  
we compute the **sequence of matrices** $\Psi_h^{(1)}$, which represent the response at horizon $h$:

$$
\Psi_h^{(1)} = A_{comp}^h B_{comp}^{(1)}, \quad h = 0, 1, 2, \dots, H.
$$

- $\Psi_0^{(1)} = B_{comp}^{(1)}$ → contemporaneous (impact) response  
- $\Psi_1^{(1)} = A_{comp} B_{comp}^{(1)}$ → response one period ahead  
- $\Psi_2^{(1)} = A_{comp}^2 B_{comp}^{(1)}$ → two-period response  
- and so on.

Each $\Psi_h^{(1)}$ is a $(mp \times 1)$ vector, but only the **first $m$ elements** correspond to the actual observable variables $Y_t$;  
the remaining rows correspond to lagged states and are therefore ignored in the plots of IRFs.

---

### Recursive Computation of the IRFs

The impulse responses can be computed recursively using the following relation:

$$
\Psi_0^{(1)} = B_{comp}^{(1)}, \\
\Psi_h^{(1)} = A_{comp} \Psi_{h-1}^{(1)}, \quad h = 1, 2, \dots, H.
$$

This recursive formulation is convenient in code implementation because it avoids repeatedly computing powers of $A_{comp}$.

In code, this is implemented as:

```python
    irf = np.zeros((horizon+1, m))
    irf[0, :] = b1       # Impact at horizon 0. As we can see it is simply b1

    # Iterate forward using the companion matrix
    for h in range(1, horizon+1):
        state = A_comp @ state
        irf[h, :] = state[:m, 0]   #This is the dynamic propagation of the shock over time. 
```

At each iteration of the recursion, the state vector is updated by left-multiplying it by the companion matrix.

After one iteration:
$$
\text{state} = A_{\text{comp}} \, B_{\text{comp}}^{(1)}.
$$

After two iterations:
$$
\text{state} = A_{\text{comp}}^{2} \, B_{\text{comp}}^{(1)}.
$$

$\vdots$

After $h$ iterations:
$$
\text{state} = A_{\text{comp}}^{h} \, B_{\text{comp}}^{(1)}.
$$

This recursion therefore computes the sequence of impulse responses
$\{ A_{\text{comp}}^{h} B_{\text{comp}}^{(1)} \}_{h=0}^{H}$ without explicitly forming matrix powers.

---

### Interpretation of the Results

For each horizon $h$:

- The first entry of $\Psi_h^{(1)}$ represents the **response of the policy variable** (e.g., the 2-year Treasury yield) to its own shock.  
- The remaining entries capture the **dynamic transmission** of that shock to other variables —  
  industrial production, prices, and the excess bond premium — through the VAR dynamics encoded in $A_{comp}$.

The collection of $\{\Psi_h^{(1)}\}_{h=0}^H$ forms the **Impulse Response Function** (IRF) for the identified shock.

---

### Compact Formula for IRFs

In matrix form, the IRFs for the identified structural shock can be summarized as:

$$
\boxed{
\text{IRF}_h = S \, A_{comp}^h B_{comp}^{(1)}, \quad h = 0, 1, \dots, H,
}
$$

where $S$ is a **selection matrix** that extracts the first $m$ rows of the companion system (i.e., the actual variables in $Y_t$).


---

````python
def irf_from_companion(A_comp, b1, horizon, m, p):   #Function to compute IRFs
    """
    Computes IRFs from an already-built companion matrix A_comp.
    
    A_comp : companion matrix of size (m*p × m*p)
    b1     : structural impact vector (length m)
    horizon: number of periods for the IRF
    m      : number of variables
    p      : number of lags
    """

    # Initial state vector size: m*p
    state = np.zeros((m*p, 1)) 

    # Shock at time 0 enters in the first m positions (current variables)
    state[:m, 0] = b1

    # Store IRFs for horizon h
    irf = np.zeros((horizon+1, m))
    irf[0, :] = b1

    # Iterate forward using the companion matrix
    for h in range(1, horizon+1):
        state = A_comp @ state
        irf[h, :] = state[:m, 0]

    return irf
````

# Inference: The Wild Bootstrap

To asses the statistical significance of the estimated Impulse Response Functions (IRFs), we compute confidence intervals using a **bootstrap procedure**.

Standard errors are calculated using a **Wild Bootstrap**.
**Why?** Economic data often has "Heteroskedasticity" (volatility changes over time). A standard bootstrap (shuffling residuals) destroys this pattern. A Wild Bootstrap keeps the residuals in place but flips their signs randomly.

**Algorithm:**
1.  Get original residuals $\hat{u}_t$ and proxy $z_t$.
2.  Generate a random variable $w_t$ that is either $+1$ or $-1$ (probability 0.5).
3.  Create bootstrap residuals $u^*_t = \hat{u}_t \times w_t$.
4.  Create bootstrap proxy $z^*_t = z_t \times w_t$. **(Crucial: Flip both to preserve correlation)**.
5.  Build artificial data $Y^*_t$ using the estimated coefficients and $u^*_t$.
6.  Re-run the entire VAR + Proxy Identification on this artificial data.
7.  Repeat 1,000 times to get confidence bands (68% and 95%).

````python
import numpy as np
import pandas as pd

# ------------------------------------------------------------
# SETTINGS
# ------------------------------------------------------------
B = 2000
horizon = 50

T_full = Y.shape[0]        # total observations
T_eff = residuals_full.shape[0]  # T - p

irf_boot_z     = np.zeros((B, horizon + 1, m))
irf_boot_zorth = np.zeros((B, horizon + 1, m))

np.random.seed(123)

# ------------------------------------------------------------
# BOOTSTRAP LOOP
# ------------------------------------------------------------
for b in range(B):

    # 1) Wild bootstrap of VAR residuals
    eta = np.random.choice([-1, 1], size=T_eff)
    u_star = residuals_full * eta[:, None]

    # 2) Simulate Y*
    Y_star = np.zeros_like(Y)
    Y_star[:p, :] = Y[:p, :]

    for t in range(p, T_full):
        lags = []
        for lag in range(1, p + 1):
            lags.append(Y_star[t - lag])
        X_t = np.concatenate([np.ones(1), np.concatenate(lags)])
        Y_star[t, :] = X_t @ PHI + u_star[t - p, :]

    # 3) Re-estimate VAR on Y*
    X_star = []
    for i in range(p, T_full):
        lags = []
        for lag in range(1, p + 1):
            lags.append(Y_star[i - lag])
        X_star.append(np.concatenate([np.ones(1), np.concatenate(lags)]))

    X_star = np.array(X_star)
    Y_star_target = Y_star[p:, :]

    PHI_star = np.linalg.lstsq(X_star, Y_star_target, rcond=None)[0]
    residuals_star = Y_star_target - X_star @ PHI_star #The new residuals are now from 1973-2020 but our proxy goes from 1988-2020 so we need to trim again these new residuals 

    # 4) TRIM residuals AND eta to proxy sample. We multiply eta to get correct signs also for the proxy variables that need to be aligned with residuals (u_star)
    residuals_star_df = pd.DataFrame(
        residuals_star,
        index=df_original.index[p:],
        columns=residual_cols
    )

    eta_df = pd.DataFrame(
        eta,
        index=df_original.index[p:],
        columns=["eta"]
    )

    merged_star = (
        residuals_star_df
        .join(proxy_df, how="inner")
        .join(eta_df, how="inner")
    )

# --- TRIMMING RESIDUALS AND ETA
    u_star_trim = merged_star[residual_cols].values
    eta_trim    = merged_star["eta"].values

# --- TO PRESERVE IDENTIFICATION COVARIANCE:
    z_trim     = merged_star["MPS"].values * eta_trim
    zorth_trim = merged_star["MPS_ORTH"].values * eta_trim



    # 5) Proxy identification
    b1_star_z     = proxy_svar_identification(u_star_trim, z_trim)
    b1_star_zorth = proxy_svar_identification(u_star_trim, zorth_trim)
    
    #🟠 5.1 IMPORTANT NORMALIZING INSIDE THE BOOTSTRAP LOOP

    target_shock = 0.25   # 25 basis points --- it is impotant to normalize inside the bootstrap loop to have correct confidence intervals

    # Normalization for b1_z
    scale_z = target_shock / b1_star_z[0]
    b1_star_z_scaled = b1_star_z * scale_z

    # Normalization for b1_z_orth
    scale_zorth = target_shock / b1_star_zorth[0]
    b1_star_zorth_scaled = b1_star_zorth * scale_zorth


    # 6) Companion matrix
    coef_only_star = PHI_star[1:, :]
    top_block_star = coef_only_star.T

    bottom_left  = np.eye(m * (p - 1))
    bottom_right = np.zeros((m * (p - 1), m))
    bottom_block = np.hstack([bottom_left, bottom_right])

    A_comp_star = np.vstack([top_block_star, bottom_block])

    # 7) IRFs NOW SCALED✅  #We show the scaled IRFs directly
    irf_boot_z[b] = irf_from_companion(
        A_comp_star, b1_star_z_scaled, horizon, m, p
    )

    irf_boot_zorth[b] = irf_from_companion(
        A_comp_star, b1_star_zorth_scaled, horizon, m, p
    )
````
